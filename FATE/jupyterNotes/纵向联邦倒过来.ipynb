```python
from fate.arch.dataframe import PandasReader
import pandas as pd
from utils.FateUtils import *
from fate.ml.ensemble.algo.secureboost.hetero.guest import HeteroSecureBoostGuest
from fate.ml.ensemble.algo.secureboost.hetero.host import HeteroSecureBoostHost
```

```python
A_df = pd.read_csv(A_host_path)
A_df['sample_id'] = [i for i in range(len(A_df))]

B_df = pd.read_csv(B_guest_path)
B_df['sample_id'] = [i for i in range(len(B_df))]
```

```python
A_df.head()
```

```python
B_df.head()
```

```python
A_reader = PandasReader(sample_id_name='sample_id', match_id_name='id',dtype='float32')
```

```python
B_reader = PandasReader(sample_id_name='sample_id', match_id_name='id',label_name='y',dtype='float32')
```

```python
A_host_ctx, B_guest_ctx = create_host_guest_ctx()
```

```python
A_data = A_reader.to_frame(A_host_ctx, A_df)
```

```python
B_data = B_reader.to_frame(B_guest_ctx, B_df)
```

```python
A_host_bst = HeteroSecureBoostHost(num_trees=3, max_depth=3)
```

```python
B_guest_bst = HeteroSecureBoostGuest(num_trees=3, objective='binary:bce', max_depth=3, learning_rate=0.3)
```

```python
A_host_bst.fit(A_host_ctx, A_data)
```

```python
B_guest_bst.fit(B_guest_bst,B_data)
```
